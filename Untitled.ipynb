```python
#Import Dependendencies

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Importing our data into pandas from csv

Toronto_data = 'Bike Share Toronto Ridership_Q1 2018.csv'
bike_trips_df = pd.read_csv(Toronto_data, low_memory = False)
bike_trips_df.head()
```

```python
#Splitting data based upon 'user_type'

User_split = bike_trips_df.groupby('user_type')

#Discover on User type took bike trips

User_trips = User_split["user_type"].count()

User_trips.head()
```

```python
# Charting Data

User_chart = User_trips.plot(kind="bar", title = "Bike trips by user type")
User_chart.set_xlabel("User Type")
User_chart.set_ylabel("Number of trips taken")
plt.show()
```

```python
#Splitting data in terms upon 'trip ID' & 'User type'

biking_groups = bike_trips_df.groupby(['trip_id','user_type'])
biking_groups.head()

# Creating a new variable that holds sum of our group

sum_it_up = biking_groups.sum()
sum_it_up.head(20)
```

```python
#Splitting data in terms upon 'station ID' & 'User type'
Station_groups = bike_trips_df.groupby(['from_station_id','user_type'])
Station_groups.count()


#Creating a new variable that holds sum of our groups
sum_it_up = Station_groups.sum()

#Deleting extra columns

del sum_it_up["trip_id"]
del sum_it_up["to_station_id"]

sum_it_up.head(20)
```

```python
from_station_id = 7000

just_one_station =sum_it_up.loc[from_station_id]

User_list = just_one_station.keys()

Station_pie = just_one_station.plot(kind="pie", y= User_list, title=("Trips" + " from_station_id"))
Station_pie.set_ylabel("Total Trip Duration")

plt.show()
plt.tight_layout()
```

```python
#renaming the columns
renaming_df=bike_trips_df.rename(columns={"from_station_id":"Station ID"})

#Keeping useful data and sizing(count function) to calculate number of rides and adding the index
df = renaming_df.groupby(['Station ID', 'from_station_name']).size().reset_index(name="Number of Rides")
df.head()
```

```python
#Maximum number of rides taken from a particular station
df["Number of Rides"].max()
```

```python
#Minimum number of rides taken from a particular station
df["Number of Rides"].min()
```

```python
#Mean of number of rides taken from the data frame
df["Number of Rides"].mean()
```

```python
#Bottom 20 stations ( Prediction on Number of rides less then 96)

Bottom_20_stations = df.loc[df["Number of Rides"]<96,:]

Bottom_20_stations.head(20)
```

```python
#Bar graph showing Bottom 20 stations (Station Names vs Number of Rides)

Station = Bottom_20_stations["from_station_name"]

Rides = Bottom_20_stations["Number of Rides"]
#len(Station)

x_axis=np.arange(1,len(Station)+1)
#x_axis
plt.bar(x_axis,Rides,color="b",align="center")
plt.title("Station Names vs Number of Rides")
plt.xlabel("Station Names")
plt.ylabel("Number of Rides")
plt.xticks(x_axis,Station, rotation="vertical")
```

```python
#Line graph showing Bottom 20 stations (Station ID vs Number of Rides)
Bottom_20_stations.plot("Station ID","Number of Rides")
plt.xlim(7030,7360)
plt.ylim(-10,100)
plt.title("Station ID vs Number of Rides")
plt.xlabel("Station ID")
plt.ylabel("Number of Rides")
```

```python
#Top 20 stations ( Prediction on Number of rides more then 1500)

Top_20_stations = df.loc[df["Number of Rides"]>1500,:]
Top_20_stations.head(20)
```

```python
#Bar graph showing Top 20 stations (Station Names vs Number of Rides)
Station = Top_20_stations["from_station_name"]

Rides = Top_20_stations["Number of Rides"]
#len(Station)

x_axis=np.arange(1,len(Station)+1)
#x_axis
plt.bar(x_axis,Rides,color="r",align="center")
plt.title("Station Names vs Number of Rides")
plt.xlabel("Station Names")
plt.ylabel("Number of Rides")
plt.xticks(x_axis,Station, rotation="vertical")
```

```python
#Line graph showing Top 20 stations (Station ID vs Number of Rides)

Top_20_stations.plot("Station ID","Number of Rides")
plt.xlim(6950,7300)
plt.ylim(1000,3000)
plt.title("Station ID vs Number of Rides")
plt.xlabel("Station ID")
plt.ylabel("Number of Rides")
```
